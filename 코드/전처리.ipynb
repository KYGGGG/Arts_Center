# 데이터 합치기

## 할인코드로 할인타입범주, 할인 % 합치기

```python
import pandas as pd

df1 = pd.read_csv('opera.csv', encoding = 'UTF8')
df2 = pd.read_csv('discount_type.csv', encoding = 'UTF8')
```

```python
df1.head()
```

```python
df2.head()
```

```python
opera_1 = pd.merge(df1, df2)
```

```python
opera_1['disc_t'].value_counts()
```

```python
opera_1['disc_p'].value_counts()
```

```python
opera_1[['price','discount_type', 'disc_t','disc_p']].sample(10)
```


## 가격이 0원이면서 할인 타입이 초대, 일반인 경우

```python
opera_1.loc[opera_1['disc_t'] == '초대', ['price']].value_counts() # 초대권 받은사람 중 결제한 사람 3명
```

```python
opera_1.loc[opera_1['disc_t'] == '초대', ['price']].value_counts() # 초대권 받은사람 중 결제한 사람 3명
```

```python
opera_1.loc[opera_1['disc_t'] == '기획사', ['price']].value_counts() # 기획사 판매
```

```python
opera_1['disc_t'].value_counts()
```
