# 할인 종류 전처리

```python
import pandas as pd

df = pd.read_csv('할인 종류.csv', encoding = 'cp949')
df.drop(['Unnamed: 0'], axis = 1, inplace = True)
df.columns = ['discount_type']
df['disc_t'] = '기타'         # 할인 종류 범주화 칼럼
df['disc_p'] = '확인필요'     # 할인율 칼럼
```

```python
df['discount_type'].sort_values().unique()
```

## 할인율

```python
# 할인율 추출

dis_p_index = df[df['discount_type'].str.contains('%')].index
df.iloc[dis_p_index,2] = df.iloc[dis_p_index,0].str[-3:]
```

```python
df['disc_p'].unique()
```

```python
df[df['disc_p'] == ')5%']
```

```python
# 할인율이 5%인 내용 정리

five_p = ['인5%', '\)5%'] # \)인 이유 : 괄호가 중복되어 오류 발생

for s in five_p :
  dis_p_index_5 = df[df['discount_type'].str.contains(s)].index
  df.iloc[dis_p_index_5,2] = '5%'

df[df['disc_p'] == '5%']
```

```python
# '0%_' 정리

dis_p_index_ = df[df['disc_p'] == '0%_'].index
df.iloc[dis_p_index_,2] = df.iloc[dis_p_index_,0].str[-4:-1]

df.iloc[dis_p_index_,:]
```

```python
# '%할인' 정리

dis_p_index_dis = df[df['disc_p'] == '%할인'].index
df.iloc[dis_p_index_dis,2] = df.iloc[dis_p_index_dis,0].str[-5:-2]
df.iloc[dis_p_index_dis,:]
```

```python
df['disc_p'].unique()
```

```python
df['disc_p'].value_counts()
```

## 할인 종류

```python
df['discount_type'].sort_values().unique()
```

```python
# 할인 종류 딕셔너리
# 문화릴레이 이전에 공연을 관림 시 다음 공연 할인
disc_dict = {'경로' : '경로우대',
             '실버' : '경로우대',
             '65세' : '경로우대',
             '기획사' : '기획사',
             '초대' : '초대',
             '노블회원' : '멤버쉽',
             '블루회원' : '멤버쉽',
             '싹틔우미' : '멤버쉽',
             '골드회원' : '멤버쉽',
             '그린회원' : '멤버쉽',
             '코로나' : '코로나',
             '후원' : '후원',
             '장애인' : '장애인 및 국가유공자',
             '유공자' : '장애인 및 국가유공자',
             '초,중,고,' : '학생',
             '초/중/고/' : '학생',
             '대학' : '학생',
             '청소년' : '학생',
             '릴레이' : '릴레이',
             '조기' : '조기예매',
             '가족' : '가족',
             '패밀리' : '가족',
             '얼리버드' : '조기예매',
             '카드' : '카드'}
```

```python
# 할인 종류 범주화

for keys, values in disc_dict.items() :
  dis_t_index = df[df['discount_type'].str.contains(keys)].index
  df.iloc[dis_t_index,1] = values
```

```python
df['disc_t'].value_counts()
```

```python
df.to_csv('discount_type.csv', index = None)
```
